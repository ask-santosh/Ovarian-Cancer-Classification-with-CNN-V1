```python
train_dir = 'train_thumbnails/'
meta_data = pd.read_csv("train.csv")
# meta_data['path'] = base_image_dir + meta_data['image_id'].astype(str) + '.png'
meta_data['path_thumb'] = train_dir + meta_data['image_id'].astype(str) + '_thumbnail.png'
meta_data.head()
```

```python
from skimage.io import imread
from skimage import io

labels = ['HGSC', 'LGSC', 'EC', 'CC', 'MC']
n_samples = 5

# Assuming n_samples is already defined somewhere in your code
fig, axs = plt.subplots(len(labels), n_samples, figsize=(10, 10))

for i, lab in enumerate(labels):
    img_paths = meta_data.loc[(meta_data['label'] == lab) & (meta_data['is_tma'] == False), 'path_thumb']
    
    # Limit the number of image paths to n_samples
    limited_img_paths = img_paths.iloc[:n_samples]
    
    for j, img_path in enumerate(limited_img_paths):
        ax = axs[i, j]
        img = io.imread(img_path)
        ax.imshow(img, interpolation='bilinear')

        img_id = os.path.splitext(os.path.basename(img_path))[0]
        ax.set_title(f'Image ID: {img_id}\nLabel: {lab}', fontsize=9)
        ax.axis('off')

for ax in axs.ravel():
    ax.set_aspect('auto')

plt.tight_layout()
plt.show()
```

```python
#Image data preprocessing
from PIL import Image 
import numpy as np
image_data = []
image_label = []
path = "./train_thumbnails/"
for img , label in zip(df_no_tma['image_id_path'],df_no_tma['label']):
    image = Image.open(path+img)
    image = image.resize((224,224))
    image = image.convert("RGB")
    image = np.array(image)
    image_data.append(image)
    image_label.append(label)
```

```python
print(len(image_data))
print(len(image_label))
```

```python
set(image_label)
```

```python
img_label_ = []
for i in image_label:
    if i=="CC":
        img_label_.append(0)
    elif i=="EC":
        img_label_.append(1)
    elif i=="HGSC":
        img_label_.append(2)
    elif i=="LGSC":
        img_label_.append(3)
    elif i=="MC":
        img_label_.append(4)
```

```python
len(img_label_)
```

```python
x = np.array(image_data)
y = np.array(img_label_)
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test, y_train, y_test = train_test_split(x,y,test_size=0.15,shuffle=True)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train_scaled = x_train/255
x_test_scaled = x_test/255
# x_train_scaled
```

```python
from sklearn.metrics import classification_report , confusion_matrix , accuracy_score , auc
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Input, Dense,Conv2D , MaxPooling2D, Flatten,BatchNormalization,Dropout
from tensorflow.keras.preprocessing import image_dataset_from_directory
# import tensorflow_hub as hub 

model = Sequential()

model.add(Conv2D(filters=140,kernel_size=(3,3),strides=(1,1),activation='relu',input_shape=(224,224,3)))
model.add(keras.layers.Dropout(0.2))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=120,kernel_size=(3,3),strides=(1,1),activation='relu',input_shape=(224,224,3)))
model.add(keras.layers.Dropout(0.1))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=100,kernel_size=(3,3),strides=(1,1),activation='relu'))
model.add(keras.layers.Dropout(0.15))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=80,kernel_size=(3,3),strides=(1,1),activation='relu'))
model.add(keras.layers.Dropout(0.15))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=64,kernel_size=(3,3),strides=(1,1),activation='relu',input_shape=(224,224,3)))
model.add(Dropout(0.15))
model.add(MaxPooling2D(pool_size=(2,2)))

# model.add(Conv2D(filters=32,kernel_size=(3,3),strides=(1,1),activation='relu',input_shape=(224,224,3)))
# model.add(MaxPooling2D(pool_size=(2,2)))

# model.add(Conv2D(filters=24,kernel_size=(3,3),strides=(1,1),activation='relu',input_shape=(224,224,3)))
# model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(units=600,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=600,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=5,activation='softmax'))

model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",
             metrics=["accuracy"])
model.summary()
```

```python
history = model.fit(x_train_scaled,y_train,epochs=100,
         batch_size=8,validation_data=(x_test_scaled,y_test))
```

```python
loss, acc = model.evaluate(x_test_scaled,y_test)
print("Accuracy on Test Data:",acc)
print()
loss, acc = model.evaluate(x_train_scaled,y_train)
print("Accuracy on Train Data:",acc)
```

```python
y_pred = model.predict(x_test_scaled)
y_pred_label = [np.argmax(i) for i in y_pred]
y_pred_label[:20]  ## Predicted Labels
```
